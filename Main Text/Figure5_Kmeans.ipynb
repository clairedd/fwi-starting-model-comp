```python
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
import pygmt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
sp = xr.open_dataset('CANV_SP_Kmeans_Crust.nc')
wus = xr.open_dataset('CANV_WUS_Kmeans_Crust.nc')
cs = xr.open_dataset('CANV_CS_Kmeans_Crust.nc')
```

```python
lat_min = 31.5
lat_max = 43
lon_min = -125.0
lon_max = -114.0

latitudes = np.arange(lat_min, lat_max, 0.1796875)
longitudes = np.arange(lon_min, lon_max, 0.22)

kargs = {'init': 'k-means++',
        'n_init': 50,
        'max_iter': 300
        }
```

```python
sp_array = []
profile = []

for lat in latitudes:
    for lon in longitudes:
        vsv = sp.VSV.sel(latitude = lat, longitude = lon, method='nearest')
        vsh = sp.VSH.sel(latitude = lat, longitude = lon, method='nearest')
        vav = np.sqrt(((2*(vsh)**2)+vsv**2)/3)
        vav_val = vav.values
        try:
            vav_val = np.nan_to_num(vav_val, nan = 2200.0)
            sp_array.append(vav_val)
        except:
            sp_array.append(vav_val)
        
print(len(sp_array))
```

```python
wus_array = []
profile = []

for lat in latitudes:
    for lon in longitudes:
        #print('Working on Profile for ('+str(lat)+', '+str(lon)+').')
        vsv = wus.VSV.sel(latitude = lat, longitude = lon, method='nearest')
        vsh = wus.VSH.sel(latitude = lat, longitude = lon, method='nearest')
        vav = np.sqrt(((2*(vsh)**2)+vsv**2)/3)
        vav_val = vav.values
        try:
            vav_val = np.nan_to_num(vav_val, nan = 2200.0)
            wus_array.append(vav_val)
        except:
            wus_array.append(vav_val)
        
print(len(wus_array))
```

```python
cs_array = []
profile = []

for lat in latitudes:
    for lon in longitudes:
        #print('Working on Profile for ('+str(lat)+', '+str(lon)+').')
        vsv = cs.VSV.sel(latitude = lat, longitude = lon, method='nearest')
        vsh = cs.VSH.sel(latitude = lat, longitude = lon, method='nearest')
        vav = np.sqrt(((2*(vsh)**2)+vsv**2)/3)
        vav_val = vav.values
        try:
            vav_val = np.nan_to_num(vav_val, nan = 2200.0)
            cs_array.append(vav_val)
        except:
            cs_array.append(vav_val)
        
print(len(cs_array))
```

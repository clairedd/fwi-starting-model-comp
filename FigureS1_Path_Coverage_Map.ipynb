```python
import pygmt
from salvus import namespace as sn
import numpy as np


sp = sn.Project(path="CANV_FULL")
wus = sn.Project(path="CANV256")
cs = sn.Project(path="/Users/claired/Desktop/CANV_CSEM")
```

```python
grid = pygmt.datasets.load_earth_relief(resolution="01m", region = [-127,-112,30,44.5] )
```

```python
unique_stas = []
sta_lats = []
sta_lons = []
nets = []

for event_name in sp.events.list():
    event = sp.events.get(event_name=event_name)
    for  idx in np.arange(0, len(event.receiver_name_list), 1):   
        sta = event.receivers[idx].to_json()
        args = sta['arguments']
        sncl = str(str(args['network_code'])+'.'+str(args['station_code'])+'.')
        
        if sncl not in unique_stas:
            unique_stas.append(sncl)
            
            sta_lat = (sn.simple_config.coordinate_system_utils._geocentric_to_elliptic_latitude(lat=args['latitude']))
            sta_lon = args['longitude'] 
            netcode = str(args['network_code'])
            
            sta_lats.append(sta_lat)
            sta_lons.append(sta_lon)
```

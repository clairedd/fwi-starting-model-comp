```python
from salvus import namespace as sn

wus = sn.Project(path="CANV256")
spiral = sn.Project(path = 'CANV_FULL')
csem = sn.Project(path = '/Users/claired/Desktop/CANV_CSEM')
```

```python
#CANV_SP-- 30-100 seconds
ipc = "inversion_test"
misfit_config = "tf-phase-misfit-spiral-30-100-seconds"
iterations = range(0,19)
iter_0_sp = []

for i in iterations:
    summed = 0
    simulation_config = spiral.inversions.get_simulation_name(inverse_problem_configuration= ipc, iteration_id = i)
    for event in spiral.events.list():
        try:
            misfitdict = spiral.misfits.get(simulation_config, misfit_config, event)
            vals = misfitdict.values()
            summed +=sum(vals)
        except:
            print(event+' has some issues?')
    iter_0_sp.append(summed)   
```

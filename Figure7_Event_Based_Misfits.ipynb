```python
#SPiRaL and CANV_SP

#SPiRaL
simulation_config = 'spiral_20_mesh'
misfitdict_sp_tf = sp.misfits.get(simulation_config, 'spiral_tf_wus', sp.events.list())

#CANV_SP

simulation_config = sp.inversions.get_simulation_name(inverse_problem_configuration= "inversion_1_r2", 
                                                      iteration_id = 12)
misfitdict_cnsp_tf = sp.misfits.get(simulation_config, 'cnsp_tf_wus', sp.events.list())
```

```python
#CSEM_NA and CANV_CS

#WUS256
simulation_config = 'csem'
misfitdict_cs_tf = start.misfits.get(simulation_config, 'csem_tf_wus', start.events.list())

#CANV_WUS

simulation_config = cs.inversions.get_simulation_name(inverse_problem_configuration= "inversion_2_r2", 
                                                      iteration_id = 23)
misfitdict_cncs_tf = cs.misfits.get(simulation_config, 'cncs_tf_wus', cs.events.list())
```

```python
sp_tf_rel = {}
cnsp_tf_rel = {}
wus_tf_rel = {}
cnws_tf_rel = {}
csem_tf_rel = {}
cncs_tf_rel = {}

for k, v in misfitdict_ws_tf.items():
#     if k == 'event_NEAR_COAST_OF_NORTHERN_CALIF._Mag_5.15_2021-07-18-06-46':
#         pass
#     elif k == 'event_NEAR_COAST_OF_NORTHERN_CALIF._Mag_6.21_2021-12-20-20-10':
#         pass
#     elif k == 'event_NEVADA_Mag_5.41_2020-11-13-09-13':
#         pass
#     else:
    sp_tf_rel[k] = (misfitdict_sp_tf.get(k)-v)
    cnsp_tf_rel[k] = (misfitdict_cnsp_tf.get(k)-v)
    wus_tf_rel[k] = v-v
    
    cnws_tf_rel[k] = (misfitdict_cnws_tf.get(k)-v)
    csem_tf_rel[k] = (misfitdict_cs_tf.get(k)-v)
    cncs_tf_rel[k] = (misfitdict_cncs_tf.get(k)-v)
        
        
cnsp_sorted = dict(sorted(cncs_tf_rel.items(), key=lambda item: item[1], reverse = True))

sp_tf = {}
cnsp_tf = {}
ws_tf = {}
cnws_tf = {}
csem_tf = {}
cncs_tf = {}

for k, v in cnsp_sorted.items():
    sp_tf[k] = sp_tf_rel.get(k)
    cnsp_tf[k] = cnsp_tf_rel.get(k)
    ws_tf[k] = wus_tf_rel.get(k)
    cnws_tf[k] = cnws_tf_rel.get(k)
    csem_tf[k] = csem_tf_rel.get(k)
    cncs_tf[k] = cncs_tf_rel.get(k)
    
    
print(len(sp_tf))
```
